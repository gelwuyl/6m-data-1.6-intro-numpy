## 4. Gradebook: Aggregations and broadcasting

Reuse the “Gradebook” idea from class and extend it.

1. Compute each student’s average score (per row).
2. Compute each subject’s average score (per column).
3. Create scores_centered by subtracting the subject mean from each column (broadcasting).
4. For scores_centered, compute per-student averages again.
5. Compare which student looks best by raw average vs centered average.
6. In a markdown cell, explain briefly why centered scores might give a fairer comparison.

```python
students = np.array(["S1", "S2", "S3", "S4", "S5"])
subjects = np.array(["Math", "Stats", "Python"])

scores = np.array([
    [75, 80, 85],
    [60, 65, 70],
    [90, 88, 92],
    [82, 79, 84],
    [70, 72, 78]
])

# calculate average score for each student, round to 2 decimal places
average_scores = np.round(np.mean(scores, axis=1), 2)
print("Average scores for each student:\n", average_scores)
```

```python
# each subject average score across all students, round to 2 decimal places
subject_averages = np.round(np.mean(scores, axis=0), 2)
# print the subject name along with the average score in the matrix console format
for i, subject in enumerate(subjects):
    print(f"{subject}: {subject_averages[i]}")
```

```python
# scores_centered = scores - subject_averages broadcasting to center the scores by subtracting the subject average from each score
scores_centered = scores - subject_averages
print("Scores centered by subtracting subject averages:\n", scores_centered)
```

```python
# per-student average score after centering, round to 2 decimal places
average_scores_centered = np.round(np.mean(scores_centered, axis=1), 2)
print("Average scores for each student after centering:\n", average_scores_centered)
```

Centering scores provides a fairer comparison because it accounts for the difficulty of the subject. If the "Stats" test was much harder than the "Python" test, a student's raw average might look low simply because they took a hard class. Centering shows how they performed relative to the class average for each specific subject.

## 5: Reshape & Flatten

Scenario: You have 30 days of daily website visits, and you want to summarize them by week.

1. Reshape daily_visits into a (6, 5) array representing 6 weeks × 5 days.
2. Compute total visits per week.
3. Compute average visits per day of the week (e.g., all “day 1 of week” together, all “day 2 of week” together, etc.).
4. Flatten the reshaped array back to 1D and confirm it matches the original daily_visits.

Hint: Pay attention to how reshaping groups the data; mention any assumptions you’re making about which days belong to which week.

```python
daily_visits = np.array([
    120, 130, 125, 140, 150,
    160, 170, 155, 145, 135,
    128, 132, 138, 142, 148,
    152, 158, 162, 168, 172,
    180, 190, 185, 175, 165,
    155, 145, 135, 125, 115
])

# compute total visits per week by reshaping the array to (6, 5) and summing along axis 1
weekly_visits = np.sum(daily_visits.reshape(6, 5), axis=1)
print("Total visits per week:\n", weekly_visits)
```

```python
# Reshape to (6, 5) for 6 weeks × 5 days
reshaped_visits = daily_visits.reshape(6, 5)

# Compute average visits per day of the week (across all weeks)
average_day_of_week = np.round(reshaped_visits.mean(axis=0), 2)
print("Average visits per day of the week:\n", average_day_of_week)
print("Shape of average_day_of_week:", average_day_of_week.shape)  # Should be (5,)


## why i cannot take --> average_day_of_week = np.round(weekly_visits.mean(axis=0), 2).
## weekly_visits is a 1D array of shape (6,) (the total visits for each of the 6 weeks).
## Calling .mean(axis=0) on a 1D array in NumPy computes the overall mean of all elements in the array (since there's only one axis). This returns a scalar (0D array), not a 1D array.
## np.round() on a scalar still gives a scalar. When printed, a NumPy scalar can sometimes display in a way that looks like a "1x1 matrix" (e.g., array(152.5) or similar), but it's fundamentally a single value, not an array with shape (6,) or any other shape.
## In short: You're computing the overall average of the weekly totals (a single number), not the average per day of the week (which should be 5 numbers, one per day).
```

```python
# flatten the daily visits array to 1d and compute total visits
original_weekly_visits = daily_visits.flatten()
print("back to original match?", np.array_equal(original_weekly_visits, daily_visits))
```
